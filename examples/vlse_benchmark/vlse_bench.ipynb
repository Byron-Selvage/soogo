```python
import matplotlib.pyplot as plt
import numpy as np
import pickle

from vlse_bench import myFuncs, algorithms
boundflag = "default"

myFuncStr = (
    "branin",
    "hart3",
    "hart6",
    "shekel",
    "ackley",
    "levy",
    "powell",
    "michal",
    "spheref",
    "rastr",
    "mccorm",
    "bukin6",
    "camel6",
)
myAlgorithmStr = ("SRS", "DYCORS", "CPTV", "CPTVl")

# myRfuncs = ("mccorm", "bukin6", "camel6")
# algorithms = ("SRS","DYCORS","CPTVl")

# myRfuncs = (
#     "ackley",
#     "bukin6",
#     "crossit",
#     "drop",
#     "egg",
#     "griewank",
#     "holder",
#     "levy",
#     "levy13",
#     "rastr",
# )
# algorithms = ("SRS", "DYCORS", "CPTV", "CPTVl")
# boundflag = "bounds"

nTrials = 0
optres = {}
exectime = {}
maxIter = {}
myNArgs = {}
minVal = {}
```

```python
for a in myAlgorithmStr:
    optres[a] = {}
    exectime[a] = {}
    for rFunc in myFuncStr:
        optres[a][rFunc] = []
        exectime[a][rFunc] = 0.0
        with open(
            "pickle/" + rFunc + "_" + a + "_" + boundflag + ".pkl", "rb"
        ) as f:
            (
                _,
                _,
                nTrials,
                optres[a][rFunc],
                exectime[a][rFunc],
                _,
            ) = pickle.load(f)

for rFunc in myFuncStr:
    maxIter[rFunc] = optres["DYCORS"][rFunc][0].nit
    for i in range(nTrials):
        assert maxIter[rFunc] == optres["DYCORS"][rFunc][i].nit
    myNArgs[rFunc] = len(myFuncs[rFunc].domain())
    minVal[rFunc] = myFuncs[rFunc].min()
```

```python
avgBestVals = {}
minBestPts = {}
minBestVals = {}
for rFunc in myFuncStr:
    minBestVals[rFunc] = np.inf
for a in myAlgorithmStr:
    avgBestVals[a] = {}
    for rFunc in myFuncStr:
        aux = np.array(
            [
                np.minimum.accumulate(
                    optres[a][rFunc][i].fsample
                ).tolist()
                for i in range(nTrials)
            ]
        )
        avgBestVals[a][rFunc] = np.mean(aux, axis=0)
        
        idx = np.argmin(aux)
        row_idx, col_idx = np.unravel_index(idx, aux.shape)
        if aux[row_idx, col_idx] < minBestVals[rFunc]:
            minBestPts[rFunc] = optres[a][rFunc][row_idx].sample[col_idx]
            minBestVals[rFunc] = aux[row_idx, col_idx]

for rFunc in myFuncStr:
    print(f"Problem {rFunc}")
    print(f"Minimum value: {minVal[rFunc]}")
    print(f"Min found value: {minBestVals[rFunc]} at {minBestPts[rFunc]}")
    for a in myAlgorithmStr:
        print("Avg best value for " + a + ": " + str(avgBestVals[a][rFunc][-1]))
        plt.plot(avgBestVals[a][rFunc], label=a)
    plt.legend()
    plt.show()
```

```python
rtau = 1e-2
atau = 1e-4

lpa = {}
for a in myAlgorithmStr:
    lpa[a] = {}
    for rFunc in myFuncStr:
        count = 0
        if minVal[rFunc] == 0:
            while count < len(avgBestVals[a][rFunc]):
                if avgBestVals[a][rFunc][count] - minVal[rFunc] <= atau:
                    break
                else:
                    count += 1
        else:
            while count < len(avgBestVals[a][rFunc]):
                if avgBestVals[a][rFunc][count] - minVal[rFunc] <= rtau * abs(minVal[rFunc]):
                    break
                else:
                    count += 1

        lpa[a][rFunc] = count + 1

minlpa = {}
for rFunc in myFuncStr:
    minlpa[rFunc] = np.min([lpa[a][rFunc] for a in myAlgorithmStr])

rpa = {}
for a in myAlgorithmStr:
    rpa[a] = {}
    for rFunc in myFuncStr:
        if lpa[a][rFunc] <= maxIter[rFunc]:
            rpa[a][rFunc] = lpa[a][rFunc] / minlpa[rFunc]
        else:
            rpa[a][rFunc] = np.inf

nAlpha = 100
alpha = np.linspace(1.0, 4.0, nAlpha)
rho = {}
for a in myAlgorithmStr:
    rho[a] = np.zeros(nAlpha)
    for k in range(len(alpha)):
        count = 0
        for rFunc in myFuncStr:
            if rpa[a][rFunc] <= alpha[k]:
                count += 1
        rho[a][k] = count / len(myFuncStr)

maxKappa = 98
delta = {}
for a in myAlgorithmStr:
    delta[a] = np.zeros(maxKappa + 1)
    for k in range(1, maxKappa + 1):
        count = 0
        for rFunc in myFuncStr:
            if lpa[a][rFunc] / (myNArgs[rFunc] + 1) <= k:
                count += 1
        delta[a][k] = count / len(myFuncStr)
```

```python
iter = 0
deltaStep = 0.000
for a in myAlgorithmStr:
    plt.step(alpha, rho[a]+deltaStep*iter, where="post", label=a)
    iter += 1
plt.xlabel(r"Performance ratio, $\alpha$")
plt.ylabel(r"$\rho_a(\alpha)$")
plt.legend()
plt.show()
```

```python
iter = 0
deltaStep = 0.000
for a in myAlgorithmStr:
    plt.step(np.arange(0, maxKappa + 1), delta[a]+deltaStep*iter, where="post", label=a)
    iter += 1
plt.xlabel(r"Number of simplex gradients, $\kappa$")
plt.xticks(range(0, 101, 5))
plt.ylabel(r"$\delta_a(\kappa)$")
plt.legend()
plt.show()
```
