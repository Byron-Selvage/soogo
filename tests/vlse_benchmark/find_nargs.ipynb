```python
import os

rFuncInputs = {}
for filename in os.listdir('.'):
    if filename.endswith('.r') and filename not in ['test.r', 'test2.r']:
        with open(filename, 'r') as f:
            count = 0
            rFuncInputs[filename] = ""
            for line in f:
                if "INPUT" in line or count > 0:
                    rFuncInputs[filename]+=line
                    count += 1
                    if "#####" in line:
                        break

for filename in rFuncInputs:
    print(filename)
    print(rFuncInputs[filename])
```
